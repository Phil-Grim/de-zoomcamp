```python
import dlt
import duckdb
import requests

stop_id = '490007533E'
line_id = '78'

# https://api.tfl.gov.uk/swagger/ui/index.html?url=/swagger/docs/v1#!/Line/Line_Timetable
# returns the timetable for a specified station on the given line 
res = requests.get(f'https://api.tfl.gov.uk/Line/{line_id}/Timetable/{stop_id}')
result = [res.json()]
result

pipeline = dlt.pipeline(destination='duckdb', dataset_name='bus_rides')

# run the pipeline with default settings, and capture the outcome
info = pipeline.run(result,
										table_name="rides",
										write_disposition="replace",
                    )

# show the outcome

conn = duckdb.connect(f"{pipeline.pipeline_name}.duckdb")

# let's see the tables
conn.sql(f"SET search_path = '{pipeline.dataset_name}'")
print('Loaded tables: ')
display(conn.sql("show tables"))
```

```python

print("\n\n\n Rides table below: Note the times are properly typed")
rides = conn.sql("SELECT * FROM rides__timetable__routes__schedules").df()
display(rides)

# vcucUemfFljNyQ
```
